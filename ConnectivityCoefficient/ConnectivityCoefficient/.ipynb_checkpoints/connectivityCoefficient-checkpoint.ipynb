```python
import pandas as pd
df = pd.read_csv('../Data/output.csv', header = None)
df
```

```python
dic = {}
for index in range(df.shape[0]):
    t = eval(df[1][index][8: -1])
    startNode = eval(df[0][index][6:]) + 1
    endNode = eval(df[3][index][8:])
    nowDistance = df[5][index]
    #print(nowDistance)
    dic.setdefault(t, {}).setdefault(startNode, {}).setdefault(endNode, []).append(nowDistance)
dic
```

```python
class MyGraph: #{time1: {startNode1: {endNode1: [distence1, distence2, ... ], endNode2: [distence1, distence2, ... ]}, startNode2: {}}, time2: {}}
    __G = {}
    __checked = {}
    __connectedBranch = []
    __averageShortedPathLength = 0
    __connectedBranchList = []
    
    def __init__(self, G):
        self.__G = G
        for v in self.__G:
            self.__checked.setdefault(v, 0)
            
    def __dfs(self, v):
        self.__checked[v] = 1
        self.__connectedBranch.append(v)
        for w in self.__G[v]:
            if (self.__checked[w] == 0):
                self.__dfs(w)
                
    def DFS(self):
        for v in self.__G:
            if (self.__checked[v] == 0):
                self.__dfs(v)
                self.__connectedBranchList.append(self.__connectedBranch)
                self.__connectedBranch = []
                
    def getConnectedBranchList(self):
        self.DFS()
        return self.__connectedBranchList
    
    def getAverageShortedPathLength(self):
        sum = 0
        for v in self.__G:
            for w in self.__G[v]:
                sum += min(self.__G[v][w])
        n = len(self.__G)
        self.__averageShortedPathLength = sum /(n * (n - 1))
        return self.__averageShortedPathLength
    
    def getConnectivityCoefficient(self):
        res = 0
        for index in self.getConnectedBranchList():
            res += len(index) * self.getAverageShortedPathLength() / len(self.__G)
        res *= len(self.__connectedBranchList) #æ•°å€¼åˆ†æž
        return 1 / res
```

```python
connectivityCoefficientList = []
for i in dic:
    for j in range(1, 51):
        dic[i].setdefault(j, {})
    myDic = MyGraph(dic[i])
    #print(myDic.getConnectedBranchList())
    #print(myDic.getAverageShortedPathLength())
    #print(myDic.getConnectivityCoefficient())
    connectivityCoefficientList.append(myDic.getConnectivityCoefficient())
print(connectivityCoefficientList)
```

```python
import matplotlib.pyplot as plt
x_time = []
for i in range(4, 100, 2):
    x_time.append(i)
plt.plot(x_time, connectivityCoefficientList, color = 'red', label = 'connectivityCoefficient')

plt.show()
```
