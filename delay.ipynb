```python
import pandas as pd
df = pd.read_csv('packet2.csv',header = 0)
df[:10]
```

```python
df["type"].value_counts() 
```

```python
arr1=[]
arr=[]
arr4=[]
endtime=0
for row in df.iterrows():
    # print(row[1].type)
    if(row[1].type=="t"):
        starttime=row[1].time
        sendnode=row[1].Node
        arr=[]
        arr.append(sendnode)
    else:
        endtime=row[1].time
        receivenode=row[1].Node
        arr.append(receivenode)
        delay=endtime-starttime
        arr4.append([delay,endtime])
    #if arr not in arr1:
    arr1.append(arr)
arr4
```

```python
df4 = pd.DataFrame(arr4)
df4.rename(columns={0:"delay",1:"time"},inplace=1)
df4[:10]
```

```python
for i in range(0,49):
    print(str((i+1)*2)+'s    '+str(df4[(df4[1]<2*(i+1))&(df4[1]>2*i)][0].mean()))
```
