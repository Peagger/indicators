```python
import pandas as pd
df = pd.read_csv('routes4.csv',header=None)
df
```

```python
nodes_num=20
name_dict={}#节点对照表
for i in range(nodes_num):
    name_dict['10.1.0.'+str(i+1)]=str(i)
print(name_dict)
```

```python

dp=[]#分时间写入列表
for i in range(2,100,2):
    dp.append(df[df[1]=="+{}s".format(str(i))])
print(dp[1])
```

```python
df3= dp[1]
df3.rename(columns={0:"Node",1:"Time",2:"Destination",3:"NextHop",4:"Interface",5:"Distance"})
df3
```

```python
Accesslist=[]#各个时间段的可达矩阵列表
for i in range(0,49):
    AccessMatrix=[[0 for col in range(nodes_num)] for row in range(nodes_num)]
    for row in dp[i].itertuples():#读入邻接矩阵
        i=int(row[1])
        j=int(name_dict[row[4]])
        AccessMatrix[i][j]=1
    for k in range(nodes_num):#warshall算法求传递闭包
        for i in range(nodes_num):
            for j in range(nodes_num):
                if AccessMatrix[i][j]==1:
                    AccessMatrix[i][j]=1
                else:
                    AccessMatrix[i][j]=AccessMatrix[i][k]*AccessMatrix[k][j]
    Accesslist.append(AccessMatrix)
#print(len(Accesslist))
# print(Accesslist[8])
```

```python
def f(nodes_num,k,a,m,num):#N数取K的全排列
    global count
    if k==0:
        flag=1
        #print(a)
        for i in range(0,len(a)):
            for j in range(0,len(a)):
                if (i<j and Accesslist[num][a[i]][a[j]]==0):
                    return
        count=count+flag
    else:
        for i in range(nodes_num,k-1,-1):
            a[m]=i-1
            f(i-1,k-1,a,m+1,num)
```

```python
import math
def rate(count,nodes_num,k):#N取K全排列个数
    total=(math.factorial(nodes_num)//(math.factorial(k)*math.factorial(nodes_num-k)))
    return count/total
```

```python
for i in range(0,49):#第2i秒
    for k in [2,3,nodes_num]:#2连通，k连通，全连通
        count=0
        a=[0 for col in range(k)]
        f(nodes_num,k,a,0,i)
        print('第+{}s:{}连通率:{:.4f}%'.format((i+1)*2,k,rate(count,nodes_num,k)*100))
```
